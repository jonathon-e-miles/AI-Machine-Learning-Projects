## 1. Setup and Data Import

```python
# Task 1: Import relevant Python libraries
import pandas as pd
import numpy as np
import json 

print("Libraries (Pandas, NumPy, json) imported successfully.")

# Task 2: Import the CSV file – NSMES1988.csv into a DataFrame
file_path = 'NSMES1988.csv' 
try:
    # We will assume the file is correctly placed in the current directory.
    # The 'index_col=0' argument tells Pandas to use the first unlabelled column (the row numbers) as the DataFrame index, 
    # preventing it from being treated as a data column.
    df = pd.read_csv(file_path, index_col=0) 
    print(f"File '{file_path}' imported successfully.")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
    # Fallback for demonstration purposes if file is missing
    data = {
        'visits': [5, 1, 13, 16], 'nvisits': [0, 0, 0, 0], 'age': [6.9, 7.4, 6.6, 7.6], 
        'gender': ['male', 'female', 'female', 'male'], 'income': [2.881, 2.7478, 0.6532, np.nan]
    }
    df = pd.DataFrame(data)
    print("Using a placeholder DataFrame for demonstration.")


# Display the first few rows to confirm successful loading
print("\nFirst 5 rows of the DataFrame:")
print(df.head())
```

## 2. Preliminary Data Inspection and Cleaning Check

```python
# Task 3: Report details from physical inspection (rows, columns)
print("--- DataFrame Physical Inspection ---")
print(f"Number of Rows (Records): {df.shape[0]}")
print(f"Number of Columns (Features): {df.shape[1]}")

# Task 4: Find out if the data is clean or has missing values
print("\n--- DataFrame Structure and Cleanliness Check (df.info()) ---")
df.info()

print("\n--- Missing Value Count per Column (df.isnull().sum()) ---")
# Summing the boolean results of isnull() gives the total count of missing values (NaN) per column.
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
# We can clearly see that 'income' is the only column with missing values (NaNs).

# Task 5: Comment on data types, values, and range, specifically on age and income
print("\n--- Descriptive Statistics for Age and Income ---")
print(df[['age', 'income']].describe())

print("\n--- Comments on Data Types and Range ---")

# Comment 1: Age
# The 'age' column is a float64. The range is highly constrained (Min: 6.5, Max: 8.5).
# This is unusual for raw age data and implies the data is **coded, standardized, or aggregated** (e.g., representing age in decades or standardized Z-scores). 
# This variable must be interpreted carefully.
print("1. **Age:** Type is float64. The narrow range (6.5 to 8.5) suggests the data is **coded/standardized**, not raw age in years.")

# Comment 2: Income
# The 'income' column is a float64 and is described as "Family income in USD 10000".
# The `count` in the describe output (4222) is less than the total rows (4394), confirming the **172 missing values** found in Task 4. 
# This column is not clean and requires imputation or removal of rows before modeling.
print("2. **Income:** Type is float64 and has confirmed **172 missing values (NaNs)**. It must be addressed (e.g., by imputation) before detailed analysis.")

# Comment 3: Other Data Types
# Columns like 'gender', 'married', 'employed', etc., are categorical factors currently stored as generic `object` (string). 
# This is highly inefficient for memory and should be optimized.
print("3. **Categorical Columns:** Many columns like 'gender', 'married', and 'health' are stored as `object` (strings), which presents a major opportunity for memory optimization.")
```

### Task 7: Memory Information and Optimization Recommendation

Professor's Explanation:
We calculate the current memory usage and propose downcasting strategies to make the DataFrame smaller and faster to process.

```python
# Task 7: Perform memory information and recommend non-default data types
# Calculate and print current memory usage
memory_usage_bytes = df.memory_usage(deep=True).sum()
memory_usage_mb = memory_usage_bytes / (1024 * 1024)
print(f"\n--- Memory Optimization Analysis ---")
print(f"Current DataFrame Memory Usage: {memory_usage_mb:.4f} MB")

print("\n--- Non-Default Data Type Recommendations ---")
print("1. **Integer Columns (e.g., visits, school):** Default `int64` is overkill. Should be downcast to **`int8`** or **`int16`**.")
print("2. **Float Columns (e.g., age, income):** Default `float64` is too precise. Should be downcast to **`float32`** (single precision).")
print("3. **Object Columns (e.g., gender, health):** These categorical strings should be converted to the specialized Pandas **`category`** dtype for massive memory savings.")
```

## 4. Data Cleaning and Final Export

### Task 8: Recommended Changes Before Detailed Data Analysis

```python
# Task 8: What changes would you recommend on the dataframe before attempting a detailed data analysis?

# 1. Handle Missing Values (Imputation):
# We use the median of the existing 'income' values to fill the 172 missing NaN entries. 
# The median is preferred over the mean because it is less sensitive to extreme outliers.
median_income = df['income'].median()
df['income'].fillna(median_income, inplace=True)
print(f"\nRecommended Change 1: Missing 'income' values filled with the median: {median_income:.4f}")

# 2. Implement Memory-Optimized Type Conversion:
# Apply the recommendations from Task 7.

# Downcast Integers:
int_cols = ['visits', 'nvisits', 'ovisits', 'novisits', 'emergency', 'hospital', 'school', 'chronic', 'adl']
for col in int_cols:
    if col in df.columns:
        # Use errors='coerce' to turn anything that can't be an int into NaN (though unlikely here)
        df[col] = pd.to_numeric(df[col], errors='coerce', downcast='integer')

# Downcast Floats:
float_cols = ['age', 'income']
for col in float_cols:
    if col in df.columns:
        df[col] = df[col].astype(np.float32)

# Convert Categoricals:
cat_cols = ['gender', 'married', 'health', 'region', 'employed', 'insurance', 'medicaid']
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
        
# Final check of data types and memory usage after cleaning and optimization
print("\n--- Final Data Structure Check After Cleaning and Optimization ---")
df.info()
```
